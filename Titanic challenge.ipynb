```python
import numpy as np
import pandas as pd
```

```python
#importing datasets

path_train = r'C:\Users\Jatin Mittal\Downloads\Compressed\titanic challenge kaggle\train.csv'
path_test = r'C:\Users\Jatin Mittal\Downloads\Compressed\titanic challenge kaggle\test.csv'
df_train = pd.read_csv(path_train)
y = df_train.loc[:,'Survived']

df_train = df_train.drop(columns=['PassengerId','Survived','Ticket'])
df_test = pd.read_csv(path_test).drop(columns=['PassengerId','Ticket'])

category_col = ['Sex','Embarked']
for col in category_col:
    df_train[col] = df_train[col].astype('category')
    
for col in category_col:
    df_train[col] = df_train[col].astype('category')
```

```python
# just showing testing and training dataset
print("Training set")
display(df_train.describe())
print("Testing set")
display(df_test.describe())
df_train.head()
```

```python
# appending test and training dataset
#test offset = 891

all_data = df_train.append(df_test)
print(all_data.describe())
all_data.head()
```

```python
all_data['Name'] = all_data['Name'].map(lambda x: x.split(',')[1].split('.')[0].strip())
titles = all_data['Name'].unique()
all_data.head()
```

```python
all_data['Cabin'].fillna('U', inplace=True)
all_data['Cabin'] = all_data['Cabin'].apply(lambda x: x[0])
all_data['Cabin'].unique()
```

```python
# replacing missing values in age col with average age of all others
all_data["Age"].fillna(29.8, inplace = True)
print(all_data.describe())

# checking which value has appeared most
print(all_data['Embarked'].value_counts())

# replacing missing values in Embarked col with most frequent occurence of value
all_data["Embarked"].fillna('S', inplace = True)
print(all_data.describe())
```

```python
# checking which all col still  have null values
all_data[all_data.isnull().any(axis=1)]

# from here we will replace Fare value with average value of Pclass=3 fare
```

```python
# finding mean fare of Pclass=3
all_data.groupby(['Pclass']).mean()

all_data['Fare'].fillna(13.3, inplace = True)
```

```python
all_data.head()
```

```python
# ONE HOT ENCODING OF CATEGORICAL DATA
all_data_onehot = all_data.copy()
all_data_onehot = pd.get_dummies(all_data_onehot, columns=['Sex','Embarked','Name','Cabin'], prefix = ['Sex','Embarked','Name','Cabin'])

#all_data_onehot.head()
```

```python
# separating test and training set
test_offset = 891
X_train = all_data_onehot.iloc[:test_offset,:]
X_test = all_data_onehot.iloc[test_offset:,:]
```

```python
# splitting training set into testing and training

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X_train,y,test_size=0.3)
```

```python
type(X_test)
```

```python
# scaling data
from sklearn.preprocessing import StandardScaler

# fit only to training data i.e. find mean and dev for training data
scale = StandardScaler()
scale.fit(x_train)

# apply those transformtions to x_train and x_test data set
x_train = scale.transform(x_train)
x_test = scale.transform(x_test)
```

```python
from sklearn.metrics import accuracy_score
def find_accuracy(y_test,y_pred):
    print(accuracy_score(y_test,y_pred))
    return
```

```python
# predicting testing datafor submission

X_test = scale.transform(X_test)
test_pred = model.predict(X_test)
```
